```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("Iris.csv")
df
```

```python
df=df.drop(columns='SepalLengthCm')
df=df.drop(columns='SepalWidthCm')
```

```python
df
```

```python
plt.scatter(df.PetalLengthCm,df.PetalWidthCm)
plt.xlabel("Petal_Length")
plt.ylabel("Petal_Width")
```

```python
km=KMeans(n_clusters=2)
km
```

```python
y_predict=km.fit_predict(df[['PetalLengthCm','PetalWidthCm']])
y_predict
```

```python

df['Cluster']=y_predict
df['Cluster']
```

```python
#Separate 3 clusters into 3 dataframes
df1=df[df.Cluster==0]
df2=df[df.Cluster==1]
```

```python
plt.scatter(df1.PetalLengthCm,df1.PetalWidthCm,color='red',label='C1')
plt.scatter(df2.PetalLengthCm,df2.PetalWidthCm,color='blue',label='C2')
plt.xlabel("Petal_Length")
plt.ylabel("Petal_Width")
plt.legend()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df[['PetalLengthCm']])
```

```python
df['PetalLengthCm']=scaler.transform(df[['PetalLengthCm']])
df
```

```python
scaler.fit(df[['PetalWidthCm']])
df.Age=scaler.transform(df[['PetalWidthCm']])
df
```

```python
km=KMeans(n_clusters=2)
y_predict=km.fit_predict(df[['PetalLengthCm','PetalWidthCm']])
y_predict
df['Cluster']=y_predict
df
```

```python
df1=df[df.Cluster==0]
df2=df[df.Cluster==1]
```

```python
plt.scatter(df1.PetalLengthCm,df1.PetalWidthCm,color='red',label='C1')
plt.scatter(df2.PetalLengthCm,df2.PetalWidthCm,color='blue',label='C2')
plt.xlabel("Petal_Length")
plt.ylabel("Petal_Width")
plt.legend()
```

```python
km.cluster_centers_
# centroid / cluster mean
```

```python
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='*')
plt.scatter(df1.PetalLengthCm,df1.PetalWidthCm,color='red',label='C1')
plt.scatter(df2.PetalLengthCm,df2.PetalWidthCm,color='blue',label='C2')
plt.xlabel("Petal_Length")
plt.ylabel("Petal_Width")
plt.legend()
```

```python
Sum_of_squared_distances=[]
K=range(1,10)
for k in K:
  km=KMeans(n_clusters=k)
  km=km.fit(df[['PetalLengthCm','PetalWidthCm']])
  Sum_of_squared_distances.append(km.inertia_)
Sum_of_squared_distances
# inertia is used to get sse
```

```python
plt.plot(range(1,10),Sum_of_squared_distances)
plt.title("Elbow Method")
plt.xlabel('No. of Clusters')
plt.ylabel('Sum_of_squared_distances')
```
